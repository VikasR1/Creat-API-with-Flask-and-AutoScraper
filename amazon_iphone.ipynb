```python
from autoscraper import AutoScraper

url = 'https://www.amazon.in/s?k=iphones'
wanted_list = ["₹51,990"]

scraper = AutoScraper()
result = scraper.build(url, wanted_list)

# print(result)
scraper.get_result_similar(url,grouped=True)
print(scraper.get_result_similar(url,grouped=True))

```

```python
scraper.set_rule_aliases({'rule_oe63':'Price'})
scraper.keep_rules('rule_oe63')

scraper.save('amazone-iphone')
```
