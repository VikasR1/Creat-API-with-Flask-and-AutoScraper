# Getting similar results

Say we want to fetch all artist, artist name, label and date of release and price page:

```python
from autoscraper import AutoScraper

#refrence: https://github.com/alirezamika/autoscraper
```

```python
url = 'https://www.junodownload.com/electro/back-cat/releases/1'

wanted_list = ["_codebox", "X_y_z", "Frustrated Funk", "13 Dec 19","from ₹110.85"]

scraper = AutoScraper()
result = scraper.build(url, wanted_list)

scraper.get_result_similar(url,grouped=True)
print(scraper.get_result_similar(url,grouped=True))
```

```python
scraper.set_rule_aliases({'rule_optw':'Artist'})
scraper.keep_rules('rule_optw')

scraper.save('juno-page-details')
```
